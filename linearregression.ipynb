```python
#evlerin m2 karesine göre fiyat tahmını yapan linear regresyon modeli
import numpy as np
X=np.array([50,100,150,200,250])
y=np.array([100000,200000,300000,400000,500000])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
plt.xlabel('evin m2 si')
plt.ylabel('evin fiyatı- TL')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X.reshape(-1,1),y)
```

```python
metre_kare=int(input('lutfen tahmin edilmesi gereken evin alanını giriniz(m2 cinsinden):'))
tahmin_fiyat=model.predict([[metre_kare]])
print(f'{metre_kare} m2 lik bir evin fiyatı: {tahmin_fiyat[0]}')
```
