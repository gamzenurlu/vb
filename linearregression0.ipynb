```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
from sklearn import datasets
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
```

```python
#iris veri setinin iceriye aktarılması
iris=datasets.load_iris()
X=iris.data
y=iris.target
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
y_train
```

```python
#Decision tree ile sınıflandırma 
dt=DecisionTreeClassifier(random_state=0)
dt.fit(X_train, y_train)
dt_predict= dt.predict(X_test)
print(f"Accuracy Skor: {accuracy_score(y_test,dt_predict)}")
print(f"Precision Skor: {precision_score(y_test,dt_predict,average='weighted')}")
print(f"Recall Skor: {recall_score(y_test,dt_predict,average='weighted')}")
print(f"F1 Skor: {f1_score(y_test,dt_predict,average='weighted')}")
```
