```python
#evin m2
#evin oda sayisi
#mahallenin gelir seviyesi
import matplotlib.pyplot as plt
import numpy as np
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
np.random.seed(0)
area=np.random.uniform(100,200,50)
rooms=np.random.randint(1,5,50)
budget=np.random.uniform(20,40,50)

b0=5000
b1=300
b2=1000
b3=2000

price=b0+(b1*area)+(b2*rooms)+(b3*budget)
price
```

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(projection='3d')
ax.scatter(area,rooms,price,color='black',label='Ev fiyat',alpha=0.6)
area_grid, rooms_grid=np.meshgrid(np.linspace(100,200,10), np.linspace(1,5,10))
budget_grid=30
price_grid=b0+(b1*area_grid)+(b2*rooms_grid)+(b3*budget_grid)

ax.plot_surface(area_grid,rooms_grid,price_grid,color='lightgreen',alpha=0.5)
ax.set_xlabel('Ev alan')
ax.set_ylabel('oda says')
ax.set_zlabel('ev fiyat')
plt.legend()
plt.show()
```

```python
#linear regression
np.random.seed(0)
area=np.random.uniform(100,200,50)
rooms=np.random.randint(1,5,50)
budget=np.random.uniform(20,40,50)

#bagmsz degskenler
X=np.column_stack((area,rooms,budget))

y=b0+(b1*area)+(b2*rooms)+(b3*budget)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X)
y_pred
```

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(projection='3d')
#gercek verilerin gorseli
ax.scatter(area,rooms, y, color='blue',label='gercek ev fiyatlar',alpha=0.5)

#tahmin edilen verileirn gorseli
ax.scatter(area,rooms, y_pred, color='red',label='tahmin edilen ev fiyatlar',alpha=0.6)
ax.set_xlabel('Ev alan')
ax.set_ylabel('oda says')
ax.set_zlabel('ev fiyat')
plt.legend()
plt.show()
```

```python
print(f'modelin kat saylar: {model.coef_}')
print(f'modelib sabit terimi: {model.intercept_}')
print(f'modelin egitim skoru: {model.score(X_train, y_train)}')
print(f'modelin test skoru: {model.score(X_test, y_test)}')
```
